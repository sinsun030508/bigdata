평균과 중앙값 계산하기
비정상적으로 큰 값이나 작은 값을 지칭하여 **이상치**라고 함



```python
import numpy as np
def fjumsu(x):
 cnt = 0
 # f학점(60점 미만의 갯수를 구행서
 for i in range(len(x)):
  if x[i]<60:
      cnt=cnt+1
 fmax = np.max(x)
 return cnt, fmax

jumsu = ([[100,80,60,70,80],
          [80,90,50,60,100]])

cnt1, max1 = fjumsu(jumsu[0])
print(f"1반의 f학점의 인원수 :  {cnt1}")
print(np.sort(jumsu[0]))
print("1반의 최고 점수 : ", max1)

cnt2, max2 = fjumsu(jumsu[1])
print(f"2반의 f학점의 인원수 :  {cnt2}")
print(np.sort(jumsu[1]))
print("2반의 최고 점수 : ", max2)
```

```python
group_A = np.array([60,70,80,90,100])
group_B = np.array([76,78,80,82,84])

print("A 모둠 학생들의 수학 점수 평균: ", np.mean(group_A))
print("B 모둠 학생들의 수학 점수 평균: ", np.mean(group_B))
```

```python
print("A 모둠의 분산:", np.var(group_A))
print("B 모둠의 분산:", np.var(group_B))

print("A 모둠의 학생의 수학점수 표준편차:", np.std(group_A))
print("B 모둠의 학생의 수학점수 표준편차:", np.std(group_B))


print("A 모둠의 학생의 수학점수 제곱근:", np.sqrt(np.var(group_A)))

print("b 모둠의 학생의 수학점수 제곱근:", np.sqrt(np.var(group_B)))
```

```python
def train_test_split(x):
  shuffled_x = np.random.permutation(len(x))
  split_index=int(len(shuffled_x)*0.8)
  train_data=shuffled_x[:split_index]
  test_data=shuffled_x[split_index:]
  return train_data, test_data


x= np.arange(1,51)
train_data, test_data = train_test_split(x)
print(f"train_data(80%) = \n{train_data}")
print(f"test_data(20%) = \n{test_data}")
```
