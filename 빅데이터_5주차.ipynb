```python
c = np.array([35,24,55,69,19,99])
c.sort()
print(c)

d = c[::-1]
print(d)

c_flip = np.flip(c)
print(c_flip)
```

```python
e = np.array([[35,24,55],[69,19,9],[4,1,11]])
e.sort()
print('--- flip() 이전 ----')
print(e)
g = np.flip(e, axis=0)
print('--- flip(e, axis=0) 이후 ----')
print(g)
h = np.flip(e, axis=1)
print('--- flip(e, axis=1) 이후 ----')
print(h)
```
